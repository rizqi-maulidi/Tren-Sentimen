##Melakukan Cleaning Text

```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords

# Download stopwords jika belum diunduh
nltk.download('stopwords')

# Fungsi untuk membersihkan URL dari teks
def remove_URL(text):
    url = re.compile(r'https?://\S+|www\.\S+')
    return url.sub(r'', text)

# Fungsi untuk membersihkan HTML dari teks
def remove_html(text):
    html = re.compile(r'<.*?>')
    return html.sub(r'', text)

# Fungsi untuk membersihkan emoji dari teks
def remove_emoji(text):
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"
        u"\U0001F300-\U0001F5FF"
        u"\U0001F680-\U0001F6FF"
        u"\U0001F1E0-\U0001F1FF"
                            "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

# Fungsi untuk membersihkan angka dari teks
def remove_numbers(text):
    text = re.sub(r'\d+', '', text)
    return text

# Fungsi untuk membersihkan simbol dari teks
def remove_symbols(text):
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  # Menghapus semua simbol
    return text

# Fungsi untuk membersihkan mention Twitter dari teks
def remove_twitter_handles(text):
    handles_pattern = re.compile(r'@\w+')  # Pattern untuk mencocokkan nama akun Twitter
    return handles_pattern.sub('', text)

# Fungsi untuk menghapus karakter aneh dari teks
def hapus_karakter_aneh(text):
    # Menghapus karakter non-ASCII
    text = text.encode('ascii', 'ignore').decode('ascii')
    # Menghapus karakter khusus menggunakan regex
    text = re.sub(r'[^\w\s]', '', text)
    return text
```

##Case Folding

```python
def case_folding(text):
  if isinstance(text, str):
    lowercase_text = text.lower()
    return lowercase_text
  else:
    return text
```

##Tokenasion & StopWord

```python
import pandas as pd
import nltk
import requests
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Pastikan untuk mengunduh tokenizer punkt dan stopwords dari nltk jika belum
nltk.download('punkt')
nltk.download('stopwords')


# Function to load custom stopwords from file
def load_custom_stopwords(url):
    response = requests.get(url)
    stopwords_list = response.text.splitlines()
    return set(stopwords_list)

# Function to remove stopwords
def remove_stopwords(text):
    if not isinstance(text, str):
        return text
    stop_words = set(stopwords.words('indonesian'))  # Menggunakan stopwords bahasa Indonesia
    custom_stopwords_url = 'https://raw.githubusercontent.com/rizqi-maulidi/UAS-Deep-Learning/main/kamusstopword.txt'
    custom_stopwords = load_custom_stopwords(custom_stopwords_url)
    stop_words.update(custom_stopwords)
    tokens = word_tokenize(text)
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    return ' '.join(filtered_tokens)

```

```python
def clean_text(text):
    text = remove_URL(text)
    text = remove_html(text)
    text = remove_emoji(text)
    text = remove_numbers(text)
    text = remove_symbols(text)
    text = remove_twitter_handles(text)
    text = hapus_karakter_aneh(text)
    text = case_folding(text)
    text = remove_stopwords(text)
    text = stem_text(text)
    return text

def case_folding(text):
    text = case_folding(text)
    return text

rentangwaktu['cleansing'] = rentangwaktu['snippet'].apply(clean_text)
rentangwaktu['cleansing'] = rentangwaktu['snippet'].apply(case_folding)
```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix, precision_recall_fscore_support, roc_auc_score, roc_curve, auc
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# File path
file_path = '/content/dataevaluasifinal.csv'

# Load the dataset
df = pd.read_csv(file_path)

y_true = df['TrueSentiment']  # Ensure this matches the correct column name in your dataset
y_pred = df['Sentiment']  # Ensure this matches the correct column name in your dataset


# Calculate confusion matrix
combined_cm = confusion_matrix(y_true, y_pred, labels=[-1, 0, 1])

# Perhitungan Akurasi dan Error Rate
total_predictions = np.sum(combined_cm)
correct_predictions = np.trace(combined_cm)
accuracy = correct_predictions / total_predictions
error_rate = 1 - accuracy

# Perhitungan Precision, Recall, dan F1 Score
precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, labels=[-1, 0, 1])

# Visualisasi confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(combined_cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Negative', 'Neutral', 'Positive'], yticklabels=['Negative', 'Neutral', 'Positive'])
plt.xlabel('Predicted Sentiment')
plt.ylabel('True Sentiment')
plt.title('Confusion Matrix')
plt.show()

print(f"Akurasi: {accuracy:.4f}")
print(f"Error Rate: {error_rate:.4f}")

# Menampilkan Precision, Recall, dan F1 Score untuk setiap kelas
for i, label in enumerate(['Negative', 'Neutral', 'Positive']):
    print(f"{label} - Precision: {precision[i]:.4f}, Recall: {recall[i]:.4f}, F1 Score: {f1[i]:.4f}")

# Menampilkan rata-rata tertimbang F1 Score
weighted_f1 = np.average(f1, weights=[sum(np.array(y_true) == -1), sum(np.array(y_true) == 0), sum(np.array(y_true) == 1)])
print(f"F1 Score Rate: {weighted_f1:.4f}")

# Encode the true and predicted labels as one-hot vectors for AUC calculation
y_true_onehot = pd.get_dummies(y_true).values
y_pred_onehot = pd.get_dummies(y_pred).values

# Calculate ROC AUC for each class
n_classes = y_true_onehot.shape[1]
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true_onehot[:, i], y_pred_onehot[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot ROC curve for each class
plt.figure(figsize=(10, 8))
for i, color in zip(range(n_classes), ['blue', 'orange', 'green']):
    plt.plot(fpr[i], tpr[i], color=color, lw=2, label=f'ROC curve (area = {roc_auc[i]:.2f}) for class {i}')
plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()

# Print AUC for each class
for i, label in enumerate(['Negative', 'Neutral', 'Positive']):
    print(f"{label} - AUC: {roc_auc[i]:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score, KFold
from sklearn.pipeline import Pipeline
from sklearn.metrics import make_scorer, accuracy_score
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

# Function to load BERT model and tokenizer
def loadBertSentimen():
    pretrained = "mdhugol/indonesia-bert-sentiment-classification"
    model = AutoModelForSequenceClassification.from_pretrained(pretrained)
    tokenizer = AutoTokenizer.from_pretrained(pretrained)
    return tokenizer, model
```

```python
# Function to perform sentiment analysis using BERT
def sentiment(text, tokenizer, model):
    sentiment_analysis = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
    label_index = {'LABEL_0': 1, 'LABEL_1': 0, 'LABEL_2': -1}
    result = sentiment_analysis(text)
    status = label_index[result[0]['label']]
    score = result[0]['score']
    return status, score
  # Custom scoring function for cross-validation
def custom_scorer(estimator, X, y):
    tokenizer, model = estimator
    predictions = []
    for text in X:
        status, _ = sentiment(text, tokenizer, model)
        predictions.append(status)
    return accuracy_score(y, predictions)
```

```python
if __name__ == '__main__':
    tokenizerS, modelS = loadBertSentimen()

    # Load your dataset (assuming it contains 'case_folding' for text and 'Sentiment' for labels)
    df = pd.read_csv('/content/dataevaluasifinal.csv')

    # Assuming 'case_folding' is your text column and 'Sentiment' is your label column
    X = df['case_folding'].values
    y = df['TrueSentiment'].values

    # Perform cross-validation
    kf = KFold(n_splits=5)
    scores = []
    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        score = custom_scorer((tokenizerS, modelS), X_test, y_test)
        scores.append(score)

    # Print cross-validation scores
    print("Cross-validation scores:", scores)
    print("Average cross-validation score:", np.mean(scores))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
import time
import joblib

# Load the data
file_path = '/content/dataevaluasifinal.csv'
df = pd.read_csv(file_path)

# Assuming the text column is named 'title' and the label column is named 'label'
text_column = 'case_folding'
label_column = 'TrueSentiment'

# Ensure the text and label columns exist in the DataFrame
if text_column not in df.columns or label_column not in df.columns:
    raise KeyError(f"Columns '{text_column}' or '{label_column}' not found in the DataFrame")

# Extract the features and labels
X = df[text_column].values
y = df[label_column].values

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=787)

# Define the pipeline
text_clf = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('clf', RandomForestClassifier(n_estimators=100, random_state=787)),
])

# Measure training time
start_time = time.time()
# Train the model
text_clf.fit(X_train, y_train)
end_time = time.time()

# Execution time
execution_time = end_time - start_time

# Predict the test data
predicted = text_clf.predict(X_test)

# Calculate training and test set accuracy
train_accuracy = text_clf.score(X_train, y_train)
test_accuracy = text_clf.score(X_test, y_test)

# Calculate precision, recall, and F1 score (macro average)
precision = metrics.precision_score(y_test, predicted, average='macro')
recall = metrics.recall_score(y_test, predicted, average='macro')
f1_score = metrics.f1_score(y_test, predicted, average='macro')

# Print the classification report
print(metrics.classification_report(y_test, predicted))

# Print the metrics
print(f"Training set accuracy: {train_accuracy * 100:.2f}%")
print(f"Test set accuracy: {test_accuracy * 100:.2f}%")
print(f"Execution time: {execution_time:.2f} seconds")
print(f"Precision (macro): {precision:.2f}")
print(f"Recall (macro): {recall:.2f}")
print(f"F1 Score (macro): {f1_score:.2f}")

# Save the trained model to a file
model_filename = 'MODELRFCOBA.pkl'
joblib.dump(text_clf, model_filename)

print(f"Model has been saved to {model_filename}")
```

```python
import pandas as pd

# Membaca data Suzuki_data dan kamus stopword
Mitsubishi_data = Mitsubishi_data  # Sesuaikan dengan nama file data Anda
with open('/content/kamusSWsidang.txt', 'r') as f:
    stopwords = set(f.read().splitlines())

# Fungsi untuk membersihkan stopwords dari teks
def remove_stopwords(text, stopwords):
    words = text.split()  # Pisahkan teks menjadi kata
    cleaned_text = [word for word in words if word.lower() not in stopwords]
    return ' '.join(cleaned_text)

# Terapkan pembersihan stopword pada kolom yang berisi teks (misalnya 'teks' di Mitsubishi_data)
Mitsubishi_data['bersih'] = Mitsubishi_data['case_folding'].apply(lambda x: remove_stopwords(x, stopwords))
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk import ngrams
from nltk.tokenize import word_tokenize
from collections import Counter

# Pastikan sudah mengimpor data Mitsubishi_data
# Filter berdasarkan sentimen
positive_data = Mitsubishi_data[Mitsubishi_data['Sentiment'] == 1]
negative_data = Mitsubishi_data[Mitsubishi_data['Sentiment'] == -1]

# Fungsi untuk membuat unigram, bigram, trigram
def generate_ngrams(text, n):
    tokens = word_tokenize(text.lower())  # Tokenisasi teks
    ngram = ngrams(tokens, n)  # Membuat ngram
    return [' '.join(gram) for gram in ngram]

# Membuat ngram untuk sentimen positif dan negatif
positive_unigrams = positive_data['bersih'].apply(lambda x: generate_ngrams(x, 1)).explode().values
negative_unigrams = negative_data['bersih'].apply(lambda x: generate_ngrams(x, 1)).explode().values

positive_bigrams = positive_data['bersih'].apply(lambda x: generate_ngrams(x, 2)).explode().values
negative_bigrams = negative_data['bersih'].apply(lambda x: generate_ngrams(x, 2)).explode().values

positive_trigrams = positive_data['bersih'].apply(lambda x: generate_ngrams(x, 3)).explode().values
negative_trigrams = negative_data['bersih'].apply(lambda x: generate_ngrams(x, 3)).explode().values

# Menghitung frekuensi ngram
positive_unigram_freq = Counter(positive_unigrams)
negative_unigram_freq = Counter(negative_unigrams)

positive_bigram_freq = Counter(positive_bigrams)
negative_bigram_freq = Counter(negative_bigrams)

positive_trigram_freq = Counter(positive_trigrams)
negative_trigram_freq = Counter(negative_trigrams)

# Membuat word cloud untuk unigram, bigram, trigram
def plot_wordcloud(freq_dict, title):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(freq_dict)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(title)
    plt.show()

# Plot wordcloud untuk masing-masing ngram
plot_wordcloud(positive_unigram_freq, "Positive Unigram Word Cloud")
plot_wordcloud(negative_unigram_freq, "Negative Unigram Word Cloud")
#plot_wordcloud(positive_bigram_freq, "Positive Bigram Word Cloud")
#plot_wordcloud(negative_bigram_freq, "Negative Bigram Word Cloud")
#plot_wordcloud(positive_trigram_freq, "Positive Trigram Word Cloud")
#plot_wordcloud(negative_trigram_freq, "Negative Trigram Word Cloud")
```
